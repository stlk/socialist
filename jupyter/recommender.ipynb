```python
import pandas as pd

import os
os.chdir('..')
os.environ.setdefault("DJANGO_SETTINGS_MODULE", "socialist.settings")
import manage
manage.read_env()
import django
django.setup()

from recommendation_engine.recommender import Recommender
```

```python
from IPython.display import display, HTML

def display_media(media):
    data = ['<p>{} - {} likes - {} distance - recommended {}<br/>{}</p><a href="{}" target="blank"><img src="{}"/></a>'.format(item['username'], item['likes'], item.get('distance', ''), item.get('recommended', ''), item['caption'], item['link'], item['url']) for item in media]
    display(HTML(''.join(data)))
```

```python
from django.contrib.auth.models import User

user = User.objects.get(id=1)
recommender = Recommender(user)
recommended_photos_all = recommender.related_photos.recommend()
len(recommended_photos_all)
```

```python
media_for_user_processed = sorted(recommended_photos_all,
                                  key=lambda m: m['comments'] * 0.01 + m['tag_weight'],
                                  reverse=True)

recommended_photos = media_for_user_processed[:50]
```

```python
recommended_user_ids = map(lambda m: m['user_id'], recommended_photos)

recommender.user_distance_data.get_user_aggregations(recommended_user_ids)
recommender.user_distance.assign_user_distances(recommended_photos)
```
